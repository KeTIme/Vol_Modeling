```python
one_day=call_df[call_df['date']=='2019-01-02']
```

```python
#example for ploting 3d
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface')
```

```python
X = np.array(one_day['moneyness'])
Y = np.array(one_day['period'])
X = X.reshape(6,13)
Y = Y.reshape(6,13)
print(X)
print(Y)
```

```python
Z = np.array(one_day['iv'])
Z = Z.reshape(6,13)
plt.figure(figsize=(15,10))

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface')
ax.set_ylabel('period')
ax.set_xlabel('moneyness')
```
