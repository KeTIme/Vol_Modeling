```python
import os, zipfile
import pandas as pd
from pandas import datetime
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
final_df=pd.read_csv("final_data_v4.csv")
final_df= final_df.set_index("Unnamed: 0")
plt.plot(final_df['0-360'])
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(final_df['10-180'], order=(0,1,1))
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

```python
cols = final_df.columns[3:43]
size = len(final_df)-int(len(final_df) * 0.8)
pred = np.zeros((size,len(cols)))
truey = np.zeros((size,len(cols)))
print (cols)
for i in range(len(cols)):
    truey[:,i] = final_df[cols[i]][int(len(final_df) * 0.8):len(final_df)]
print (truey)
```

```python
from pyramid.arima import auto_arima
for i in range(len(cols)):
    X = final_df[cols[i]].values
    size = int(len(X) * 0.8)
    train, test = X[0:size], X[size:len(X)]
    history = [x for x in train]
    stepwise_model = auto_arima(X, start_p=1, start_q=1,
                           max_p=3, max_q=3, m=12,
                           start_P=0, seasonal=False,
                           d=1, D=1, trace=True,
                           error_action='ignore',  
                           suppress_warnings=True, 
                           stepwise=True)
    col_order = stepwise_model.order
    try:
        print (cols[i])
        for t in range(len(test)):
            model = ARIMA(history, order=col_order)
            model_fit = model.fit(disp=0)
            output = model_fit.forecast()
            yhat = output[0]
            pred[t,i]=yhat
            history.append(test[t])
    except:
        for t in range(len(test)):
            model = ARIMA(history, order=(0,1,1))
            model_fit = model.fit(disp=0)
            output = model_fit.forecast()
            yhat = output[0]
            pred[t,i]=yhat
            history.append(test[t])

```

```python
print (mean_squared_error(truey[:,10], pred[:,10]))
```
